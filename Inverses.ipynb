```julia
function Inv(u::MultiVector{Float64,ℝ,2}) #1D inverse. Inverts a multivector if it can be inverted.
    a0=u[0][1]
    a1=u[1][1]
    
    A=[a0 a1
       a1 a0]
    b=zeros(2,1)
    b[1]=1
    
    d=det(A)
    if d==0
        return NaN
    else
        x=A\b
        return x[1]+x[2]v1
    end
end
```

```julia
function Inv(u::MultiVector{Float64,ℝ^2,4}) #2D inverse. 
    a0=u[0][1]
    a1=u[1][1]
    a2=u[1][2]
    a12=u[2][1]
    
    A=[  a0   a1  a2 -a12
         a1   a0 a12  -a2
         a2 -a12  a0   a1
        a12  -a2  a1   a0]
    b=zeros(4,1)
    b[1]=1
    
    d=det(A)
    if d==0
        return NaN
    else
        x=A\b
        return x[1]+x[2]v1+x[3]v2+x[4]v12
    end
end
```
