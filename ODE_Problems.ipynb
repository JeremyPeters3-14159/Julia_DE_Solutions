```julia
using DifferentialEquations
using Plots
w = 1
M = [0  1
    -w^2  0]
u0=rand(2,1)
tspan=(0.0,2*pi)
f(u,p,t)=M*u
prob = ODEProblem(f,u0,tspan)
sol=solve(prob)
plot(sol)
```

```julia
tspan=(0.0,2*pi)
function bc(residual,u,p,t)
    residual[0]=u[1][1]  #Dirichlet Boundary Conditions x=0 at t=0
    residual[1]=u[end][1]-1 #Dirichlet Boundary Conditions x=1 at t=tend
end
prob=TwoPointBVProblem(f,bc,[0.0,1/sin(1)],tspan)
sol=solve(prob)
plot(sol)
```

```julia
tspan=(0.0,pi/2)
function bc(residual,u,p,t)
    residual[0]=u[1][2] #Neumann Boundary Conditions
    residual[1]=u[end][2] #Neumann Boundary Conditions
end
prob=TwoPointBVProblem(f,bc,[0,2*pi],tspan)
sol=solve(prob)
plot(sol)
```

```julia
using ApproxFun
using LinearAlgebra
using Plots
d=(-1..1)^2
Δ=ApproxFun.Laplacian(d)
b=ones(∂(d))
u=\([Dirichlet(d);Δ+100I],[b;0];tolerance=1e-5)
Plots.plot(u)
```
