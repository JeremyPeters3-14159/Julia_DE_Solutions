```julia
using Plots, ApproxFun, LinearAlgebra
```

```julia
d=(-1..1)^2
g=Fun((x,y)->real(exp(x+im*y)),∂(d))
Δ=Laplacian(d)
u=[Dirichlet(d); Δ] \ [g; 0]
plot(u)
```

```julia
dx=dy=-1.0..1.0
d=dx×dy
x,y=Fun(∂(d))
x,y=components(x),components(y)
```

```julia
x[1]
```

```julia
y[1]
```

```julia
Δ=Laplacian(d)
```

```julia
B=[I⊗lneumann(dy); rneumann(dx)⊗I; I⊗rdirichlet(dy); ldirichlet(dx)⊗I]
```

```julia
g=[-exp(x[1])*sin(-1); exp(1)*cos(y[2]);
      exp(x[3])*cos(1);  exp(-1)*cos(y[4])]
u= \([B;Δ],[g;0];tolerance=1e-10)
```

```julia
plot(u)
```

```julia
f=Fun((x,y)->exp(-10(x+0.2)^2-20(y-0.1)^2))
d=domain(f)
Δ=Laplacian(d)
u= [Dirichlet(d);Δ]\[0;f]
plot(u)
```

```julia
dx=-1.0..1.0;dt=0.0..2.0
d = dx × dt
Dx=Derivative(d,[1,0]);Dt=Derivative(d,[0,1]);
u0=Fun(x->exp(-20*x^2))
u=\([I⊗ldirichlet(dt); ldirichlet(dx)⊗I; Dt+Dx],[u0;0;0];tolerance=1e-3)
```

```julia
plot(u)
```
