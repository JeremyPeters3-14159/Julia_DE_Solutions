```julia
using ApproxFun, LinearAlgebra, Plots
```

```julia
dx=(-1.0..1.0);dt=(0.0..2.0)
d=dx×dt
Dx=Derivative(d,[1,0]); Dt=Derivative(d,[0,1])
x,t=Fun(identity,d)
u0=Fun(x->exp(-20x^2),dx)
```

```julia
@time u=\([timedirichlet(d);Dt-x*Dx],[u0,[0,0],0];tolerance=1e-3)
```

```julia
plot(u)
```

```julia
dx=(-1..1);dt=(0..2)
d=dx×dt
Dx=Derivative(d,[1,0]);Dt=Derivative(d,[0,1])
x,y=Fun(d)
u0=Fun(x->exp(-20x^2),dx)
@time u=\([I⊗ldirichlet(dt);Dt+x*Dx],[u0;0];tolerance=1e-10)
plot(u)
```

```julia
dx=(-1.0..1.0);dt=(0.0..1.0)
d=dx × dt
eps=0.01
Dx=Derivative(d,[1,0]);Dt=Derivative(d,[0,1])
x,t=Fun(d)
V=2.0+x
L=Dt-eps*Dx^2-V*Dx
u0=Fun(x->exp(-20x^2),dx)
@time u=\([timedirichlet(d);L],[u0,[0,0],0];tolerance=1e-6)
contourf(u)
```

```julia
dx=Domain(-1..1);dt=Domain(0..4)
d=dx × dt
Dx=Derivative(d,[1,0]);Dt=Derivative(d,[0,1])
DD=Dt^2-Dx^2
B=[I⊗ldirichlet(dt);I⊗lneumann(dt);ldirichlet(dx)⊗I;rneumann(dx)⊗I]
u0=Fun(x->exp(-20(x-0.1)^2),dx)
u=\([B;DD],[u0;0;0;0;0];tolerance=1e-4)
contourf(u)
```

```julia
coefficients(u)
```

```julia
dx=Domain(-1..1);dt=Domain(0..0.1)
d=dx × dt
Dx=Derivative(d,[1,0]);Dt=Derivative(d,[0,1])
L=Dt+Dx^3
B=[timedirichlet(d);rneumann(dx)⊗I]
u0=Fun(x->exp(-10x^2),dx)
u=\([B;L],[u0,[0,0],0,0];tolerance=1E-3)
contourf(u)
```

```julia
plot(u)
```

```julia
dx=Domain(-5..1);dt=Domain(0..1)
d=dx × dt
Dx=Derivative(d,[1,0]);Dt=Derivative(d,[0,1])
B=[I⊗ldirichlet(dt);Neumann(dx)⊗I;rdirichlet(dx)⊗I]
u0=Fun(x->exp(-10x^2))
u=\([B;Dt+Dx^3],[u0,[0,0],0,0];tolerance=1E-4)
plot(u)
```

```julia
dx = (0.0..1); dt = (0..0.03)
d = dx × dt
Dx=Derivative(d,[1,0]); Dt=Derivative(d,[0,1]);
x,t=Fun(d)

u0 = Fun(x->exp(-200(x-.5)^2),dx)

@time u = \([timedirichlet(d); I ⊗ lneumann(dt); Neumann(dx) ⊗ I; Dt+Dx^3],
    [u0,[0,0],0,[0,0],0]; tolerance=1E-4)

plot(u)
```
